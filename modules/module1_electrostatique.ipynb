```python
import numpy as np
import matplotlib.pyplot as plt

from core.fields import global_electric_field, global_electric_potential
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

X, Y = np.meshgrid(x, y)

charges = [(1e-9, -1, 0), (-1e-9, 1, 0)]

Ex, Ey = global_electric_field(charges, X, Y)
V = global_electric_potential(charges, X, Y)
```

```python
plt.figure(figsize=(10, 10))
plt.streamplot(x, y, Ex, Ey, density=2)
plt.contourf(X, Y, V, levels=20)
plt.contour(X, Y, V, levels=20, colors='black', linewidths=0.5)
plt.show()
```
